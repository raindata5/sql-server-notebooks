logical functions and some sys. functions

```sql
USE AdventureWorks2019
GO
SELECT
    BusinessEntityID,
    case  
    when (BusinessEntityID % 2 = 1) then 'Odd'
    else 'Even' end as [odd_even]
FROM HumanResources.Employee

```

```sql
USE AdventureWorks2019
go
SELECT
    BusinessEntityID,
    IIF((BusinessEntityID % 2) > 0 ,'Odd','Even')
from HumanResources.Employee
```

```sql
use AdventureWorks2019
go

SELECT
    SalesOrderID,
    OrderQty,
    case 
    when OrderQty < 10 then 'under 10'
    when OrderQty >= 10 AND OrderQty < 20 then '10-19'
    when OrderQty >=20 AND OrderQty < 30 then '20-29'
    when OrderQty >=30 AND OrderQty < 40 then '30-39'
    else '40 and over' END AS [ordercat]
from Sales.SalesOrderDetail ; 
```

functions in the where and order by clause

```sql
-- according to certain criteria order by  lastname or firstname
use AdventureWorks2019
GO
SELECT
    PersonType,
    FirstName,
    LastName,
    case when persontype in ('IN','SP','SC') then LastName
    else FirstName end as sorting 
from Person.Person
ORDER BY sorting
```

```sql
--another method for previous query
use AdventureWorks2019
GO
SELECT
    PersonType,
    FirstName,
    LastName 
from Person.Person
ORDER BY
    case when persontype in ('IN','SP','SC') then LastName
    else FirstName END;
```

```sql
-- method for ordering nulls in beginning
use WideWorldImporters
go
SELECT
    CityName,
    LatestRecordedPopulation,
    IIF(LatestRecordedPopulation is not null,1,0)
from Application.Cities
ORDER BY 3 ASC
```

```sql
-- method for ordering nulls at end
USE
WideWorldImporters
SELECT
    CityName,
    LatestRecordedPopulation
from Application.Cities
ORDER BY 
    case 
    when LatestRecordedPopulation is null then 2
    else 1 END ;
```

```sql
use WideWorldImporters
go
SELECT
    CityName,
    LatestRecordedPopulation
from Application.Cities
WHERE LatestRecordedPopulation is null ; 
```

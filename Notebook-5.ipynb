joins

```sql
USE AdventureWorks2019
GO
SELECT
    soh.SalesOrderID,
    sp.salesquota,
    sp.bonus
from Sales.SalesOrderHeader soh 
join Sales.SalesPerson sp on soh.SalesPersonID = sp.BusinessEntityID
```

```sql
USE AdventureWorks2019
GO
SELECT
    soh.SalesOrderID,
    sp.salesquota,
    sp.bonus,
    CONCAT(p.firstname+ ' ',middlename + ' ',lastname ) name
from Sales.SalesOrderHeader soh 
join Sales.SalesPerson sp on soh.SalesPersonID = sp.BusinessEntityID
JOIN Person.Person p on sp.BusinessEntityID = p.BusinessEntityID
```

more complex joins

```sql
USE AdventureWorks2019
GO
SELECT p.Name,
    p.ProductID,
    sod.SalesOrderID
from Production.Product p 
LEFT JOIN Sales.SalesOrderDetail sod on p.ProductID = sod.ProductID
```

```sql

-- search predicate on right table of a join usually turns into a inner join but not this case
USE AdventureWorks2019
GO
SELECT p.Name,
    p.ProductID,
    sod.SalesOrderID
from Production.Product p 
LEFT JOIN Sales.SalesOrderDetail sod on p.ProductID = sod.ProductID
WHERE sod.SalesOrderID IS NULL ;
```

```sql
-- all salesperson to salesorderheader w/ orderid even if no match
USE AdventureWorks2019
go
SELECT
    sp.*,
    soh.SalesPersonID,
    
    soh.SalesOrderID
from Sales.SalesPerson sp 
LEFT JOIN Sales.SalesOrderHeader soh on sp.BusinessEntityID = soh.SalesPersonID
```

```sql
-- all salesperson to salesorderheader w/ orderid even if no match inc. name from person.person
USE AdventureWorks2019
go
SELECT
    CONCAT(p.firstname+ ' ',middlename + ' ',lastname ) name,
    sp.*,
    soh.SalesPersonID,    
    soh.SalesOrderID
from Sales.SalesPerson sp 
LEFT JOIN Sales.SalesOrderHeader soh on sp.BusinessEntityID = soh.SalesPersonID
INNER JOIN Person.Person p on sp.BusinessEntityID = p.BusinessEntityID
```

```sql
-- all rows from salesorderheader joined to s.currencyrate & p.shipmethod
USE AdventureWorks2019
GO
SELECT
    cr.CurrencyRateID, AverageRate, soh.SalesOrderID, ShipBase
FROM Sales.SalesOrderHeader soh
LEFT JOIN Sales.CurrencyRate cr on soh.CurrencyRateID = cr.CurrencyRateID
LEFT JOIN Purchasing.ShipMethod sm on soh.ShipMethodID = sm.ShipMethodID
```

```sql
-- self join
USE WideWorldImporters
GO
SELECT
c.*,
c2.billtocustomerid
FROM Sales.Customers c
LEFT JOIN sales.Customers c2 on c.customerid = c2.billtocustomerid
WHERE c.CustomerID = 1 ;
```

```sql
-- self join to count how much billing a particular customer handles
USE WideWorldImporters
GO
SELECT
count(c.CustomerID)

FROM Sales.Customers c
LEFT JOIN sales.Customers c2 on c.customerid = c2.billtocustomerid
WHERE c.CustomerID = 1 ;
```

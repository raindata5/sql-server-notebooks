```sql
-- 1) getting a count of s.sod to see if the total count of orders is over 100,000

USE AdventureWorks2019
GO
DECLARE @Count INT ;
SELECT @Count = COUNT(*)
FROM Sales.SalesOrderDetail ;

IF @Count > 100000 BEGIN
        PRINT 'Over 100,000'
END
ELSE BEGIN
        PRINT '100,000 or less'
END ;

```

12-3 using loops with the while statement

```sql
--1) printing out the letters of the alphabet

DECLARE @num int = 65 ;
WHILE @num <= 90
    BEGIN
        PRINT CHAR(@num); -- char will change numbers to letters
        set @num += 1;
    END ;
```

```sql
-- 2) priting proudct of the two counters in the inner loop

DECLARE @a int = 1, @b int ;
WHILE @a < 101 BEGIN        
        SET @b = 1;
        WHILE @b < 6 BEGIN
                PRINT @a * @b;
                SET @b +=1;
        END
        set @a += 1;
END ;
```

```sql
-- 3) break otu of loop when outer loop counter (@a) is evenly divisible by 5
DECLARE @a int = 1, @b int ;
WHILE @a < 101 BEGIN        
        SET @b = 1;
        WHILE @b < 6 BEGIN
            IF (@a % 5) = 0 BEGIN
                PRINT 'breaking out of loop'
                BREAK;
            END ;
            PRINT @a * @b;
            SET @b +=1;
        END
        set @a += 1;
END ;
```

```sql
-- 4) declaring if number is even or odd

DECLARE @a int = 1;
WHILE @a <= 100
    BEGIN
        IF (@a % 2) = 0
            BEGIN
                PRINT 'Even'
            END;
        ELSE
            BEGIN
                PRINT 'odd'
            END;
        set @a += 1 ;
    END;
```

```sql
-- 1) populating table with customer info
USE AdventureWorks2019
GO

-- CREATE TABLE #cinfo (CustomerID int,FirstName NVARCHAR(50),LastName NVARCHAR(50), countofsales int, sumoftotaldue MONEY)
SELECT
    c.customerid,
    FirstName fname,
    LastName lname,
    COUNT(*) total_order,
    SUM(TotalDue) total_spent
into #cinfo
FROM Sales.Customer c
INNER JOIN Person.Person p on c.CustomerID = p.BusinessEntityID
INNER JOIN Sales.SalesOrderHeader soh on c.CustomerID = soh.CustomerID 
GROUP BY C.CustomerID, FirstName, LastName;
```

```sql
-- 2) table variable instead

declare @cinfo TABLE (CustomerID int,FirstName NVARCHAR(50),LastName NVARCHAR(50), countofsales int, sumoftotaldue MONEY) ;
INSERT into @cinfo (CustomerID ,FirstName ,LastName , countofsales , sumoftotaldue ) 
SELECT
    c.customerid,
FROM Sales.Customer c
INNER JOIN Person.Person p on c.CustomerID = p.BusinessEntityID
INNER JOIN Sales.SalesOrderHeader soh on c.CustomerID = soh.CustomerID 
GROUP BY C.CustomerID, FirstName, LastName;;
```

```sql
-- 3) populating a table with random numbers and then printing them
DECLARE @table1 TABLE ([index] int IDENTITY, [integer] int ) ;

DECLARE @counter1 INT = 1, @counter2 int ;
WHILE @counter1 <= 1000 BEGIN 
        INSERT INTO @table1 ([integer])
        VALUES (CAST(rand() * 10000 as int) + 1);
        SET @counter1 += 1;
    END;

SET @counter1 = 1;
WHILE @counter1 <= 1000
    BEGIN
        SELECT 
            @counter2 =[integer] /* [integer]*/
        from @table1
        WHERE [index] = @counter1 ;
        PRINT @counter2
        SET @counter1 += 1;
    END;
```
